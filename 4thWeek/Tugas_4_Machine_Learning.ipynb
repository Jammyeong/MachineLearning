Sampel kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart

```python
# Fungsi untuk membuat pie chart dari kolom kategorikal
def plot_pie_chart(column):
    data = df[column].value_counts()  # Menghitung jumlah data tiap kategori
    plt.figure(figsize=(6,6))  # Mengatur ukuran chart
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title(f'Distribution of {column}', fontsize=14)
    plt.axis('equal')  # Agar pie chart berbentuk lingkaran sempurna
    plt.show()

plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('parental level of education')
plot_pie_chart('lunch')
plot_pie_chart('test preparation course')
```

Kode sintaks dengan style color yang aesthetic

```python
def plot_pie_chart(column_name):
    # Mengambil data dan menghitung jumlah kategori
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Menggunakan palet warna estetik dari Seaborn
    colors = sns.color_palette('Set3', len(labels))  # Set3 memberikan warna yang soft dan estetik

    plt.figure(figsize=(7,7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=14)
    plt.show()

# Menampilkan pie chart untuk berbagai kolom kategorikal
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
```

Kode sintaks python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

```python
# Tentukan batas nilai minimal untuk lulus
passing_score = 75

# Buat kolom untuk menentukan apakah siswa lulus atau gagal pada masing-masing pelajaran
df['math result'] = df['math score'].apply(lambda x: 'pass' if x >= passing_score else 'fail')
df['reading result'] = df['reading score'].apply(lambda x: 'pass' if x >= passing_score else 'fail')
df['writing result'] = df['writing score'].apply(lambda x: 'pass' if x >= passing_score else 'fail')

# Buat dataframe untuk menghitung jumlah 'pass' dan 'fail' untuk setiap pelajaran
results_summary = {
    'math': df['math result'].value_counts(),
    'reading': df['reading result'].value_counts(),
    'writing': df['writing result'].value_counts()
}

# Convert to DataFrame
results_df = pd.DataFrame(results_summary)

# Visualisasi menggunakan Seaborn (Stacked Bar Chart)
results_df.T.plot(kind='bar', stacked=True, figsize=(10, 6), color=['red', 'green'])

# Tambahkan judul dan label
plt.title('Jumlah Siswa yang Lulus dan Gagal untuk Masing-masing Pelajaran', fontsize=14)
plt.xlabel('Pelajaran', fontsize=12)
plt.ylabel('Jumlah Siswa', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Hasil", loc="upper right")
plt.show()
```

insight: distribusi kelulusan dan kegagalan di setiap mata pelajaran seimbang namun kegagalan lebih mendominasi di dataset ini
